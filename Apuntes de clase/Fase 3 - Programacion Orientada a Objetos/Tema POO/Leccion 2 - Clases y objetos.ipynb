```python
class Galleta():
    chocolate=False
    def __init__(self,sabor, color):
        #
        self.sabor=sabor
        self.color=color
        print("Se acaba de crear una galleta {} y {}".format(sabor,color))
    def chocolatear(self):
        self.chocolate=True
    
    def tiene_choco(self):
        if self.chocolate:
            print("Tengo chocolate")
        else:
            print("No tengo chocolate")
```

## METODOS ESPECIALES DE LA CLASE ##

```python
class Pelicula:
    #Constructor de clase
    def __init__(self, titulo, duracion, lanzamiento, autor):
        self.titulo=titulo
        self.duracion=duracion
        self.lanzamiento=lanzamiento
        self.autor=autor
        print("Se ha creado la pelicula", self.titulo)
    
    def __del__(self):
        print("Se ha borrado la pelicula", self.titulo)
```

```python
class Pelicula:
    #Constructor de clase
    def __init__(self, titulo, duracion, lanzamiento, autor):
        self.titulo=titulo
        self.duracion=duracion
        self.lanzamiento=lanzamiento
        self.autor=autor
        print("Se ha creado la pelicula", self.titulo)
    
    # Redefinir metodo delete
    def __del__(self):
        print("Se ha borrado la pelicula", self.titulo)

    # Se redefine el metodo str
    def __str__(self):
        return "{} Lanzado en {} con una duraciion de {} minutor".format(self.titulo, self.lanzamiento, self.duracion)
    
    def __len__(self):
        return self.duracion
```

```python
p=Pelicula("El Padrino", 175, 1972, "Martin Scorcse")
```

```python
str(p)
```

```python
len(p)
```

## OBJETOS DENTRO DE OBJETOS ##

```python
class Catalogo:
    peliculas=[]

    def __init__(self, peliculas=[]):
        self.peliculas=peliculas

    def agregar(self, p):
        self.peliculas.append(p)

    def mostrar(self):
        for p in self.peliculas:
            print(p)
```

```python
p=Pelicula("El Padrino", 175, 1972, "Martin Scorcse")
```

```python
c=Catalogo([p])
```

```python
c.mostrar()
```

```python
r=Pelicula("X MEN", 175, 1990, "AUTOR")
```

```python
c.agregar(r)
```

```python
c.mostrar()
```
