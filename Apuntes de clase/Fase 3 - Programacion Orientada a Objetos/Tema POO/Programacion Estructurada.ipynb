### PROGRAMACION ESTRUCTURADA EN POO ###

```python
clientes=[
    {'Nombre': 'Hector', 'Apellido': 'Costa Guzman', 'dni': '123456'},
    {'Nombre': 'Juan', 'Apellido': 'Gonzalez Marquez', 'dni': '654321'}

]
```

```python
clientes
```

```python
def mostrar_cliente(clientes, dni):
    for c in clientes:
        if (dni == c['dni']):
            print('{}{}'.format(c['Nombre'], c['Apellido']))
            return
    print('Cliente no encontrado')



mostrar_cliente(clientes, 12345)
```

```python
mostrar_cliente(clientes,'123456')
```

```python
def borrar_cliente(clientes, dni):
    for i,c in enumerate(clientes):
        if(dni==c['dni']):
            del(c[i])
            print(str(c), " >BORRADO")
            return
    print("Usuaio no encontrado")
```

```python
borrar_cliente(clientes,'awd')
```

```python
def mostrar_todos(clientes):
    for c in clientes:
        print('NOMBRE: {} APELLIDO: {} DNI: {}' .format(c['Nombre'], c['Apellido'], c['dni']))
    return
```

```python
mostrar_todos(clientes)
```
