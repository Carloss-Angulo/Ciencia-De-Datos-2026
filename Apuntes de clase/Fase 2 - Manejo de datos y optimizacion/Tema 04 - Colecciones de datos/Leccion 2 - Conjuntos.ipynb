```python
usuarios=set()
administradores=set()

print("""
Opciones -> :
        1. Capturar usuario
        2. Capturar administrador
        3. Mostrar los datos
        4. Eliminar usuario
        5. Salir del sistema
""")


while True:
    menu_opcion=input()
    if menu_opcion=="1":
        usuario=input("Ingresa el nomnbre de usuario:")
        usuarios.add(usuario)
    elif menu_opcion=="2":
      admin=input("Ingresa el nomnbre de usuario:")
      administradores.add(admin)
    elif menu_opcion=="3":
        print(f"Usuarios en el sistema: {usuarios}")
        print(f"Administradores en el sistema: {administradores}")
    elif menu_opcion=="4":
        print("Teclea el usuario a eliminar: ")
        nombre=input()
        try:
          usuarios.remove(nombre)
          print(f"El usuario {nombre} ha sido removido del sistema")
        except KeyError:
            print(f"El usuario {nombre} no se encuentra en el sistema")
        ##Nota: discard() no hace nada si elemento no existe.
        ##      remove() lanza ecepcion
        ## el print debe llevar f para que aparezcan variables dentro del print
    elif menu_opcion=="5":
        break
    else:
        print("Opcion invalida")
```
