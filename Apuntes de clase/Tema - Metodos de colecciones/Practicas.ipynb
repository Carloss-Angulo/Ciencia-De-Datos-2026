```python
import math

class Punto:

    def __init__(self, x=0, y=0):
        self.x=x
        self.y=y

    def __str__(self):
        print("P({},{})".format(self.x, self.y))
    
    def cuadrante(self):
        
            if  self.x>0 and self.y>0:
                print("El punto P({},{}) esta en el cuadrante 1".format(self.x,self.y))
            elif self.x<0 and self.y>0:
                print("El punto P({},{}) esta en el cuadrante 2".format(self.x,self.y))
            elif self.x<0 and self.y<0:
                print("El punto P({},{}) esta en el cuadrante 3".format(self.x,self.y))
            elif self.x>0 and self.y<0:
                print("El punto P({},{}) esta en el cuadrante 4".format(self.x,self.y))
            else:
                print("El punto P({},{}) esta en el origen".format(self.x,self.y))
    def vector(self, punto):
        x= (punto.x - self.x)
        y= (punto.y - self.y)

        print("El vector entre P({},{}) y P({},{}) es AB({},{})".format(self.x,self.y,punto.x,punto.y,x,y))

    def distancia(self, punto):
        distancia = math.sqrt((punto.x - self.x)**2 + (punto.y - self.y)**2 )

        print("La distancia entre A({},{}) y B({},{}) es igual a {}".format(self.x, self.y, punto.x, punto.y, distancia))


```

```python
p=Punto(-5,4)
```

```python
p.cuadrante()
```

```python
p_1 = Punto(2,6)
p.vector(p_1)
p.distancia(p_1)
```
