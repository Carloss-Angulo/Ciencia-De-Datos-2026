# .apply Method

```python
def classify_performance(num):
    if num<300:
        return "ok"
    elif num>=300 and num <650:
        return "Satisfactory"
    else:
        return "Incredible"
```

```python
google.apply(classify_performance)
```

```python
google=pd.read_csv("Datasets/google_stock_price.csv").squeeze()
```

```python
google.apply(classify_performance).head()
```

```python
google.apply(lambda stock_price : stock_price+1)
```
