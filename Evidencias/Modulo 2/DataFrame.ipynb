# Rank Values with .rank() Method

```python
nba=pd.read_csv("Datasets/nba.csv")
```

```python
nba["Salary"]=nba["Salary"].dropna(how="all")
```

```python
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
```

```python
nba.info()
```

```python
nba["Salary"].rank(ascending=False)
```
