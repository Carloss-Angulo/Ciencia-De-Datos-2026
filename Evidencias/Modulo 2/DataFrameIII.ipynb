# The .nsmallest() and .nlargest() Methods

```python
b=pd.read_csv("Datasets/jamesbond.csv", index_col="Film")
b.sort_index(inplace=True)
b.head(3)
```

```python
b.sort_values("Box Office", ascending=False).head(3)
```

```python
b.nlargest(3,columns="Box Office")
```

```python
b.nsmallest(n=2, columns="Box Office")
```

```python
b.nsmallest(6, columns="Bond Actor Salary")
```

```python
b["Box Office"].nlargest(8)
```

```python
b["Year"].nsmallest(2) # Peliculas más antiguas
```

# The .query() Method

```python
b=pd.read_csv("Datasets/jamesbond.csv", index_col="Film")
b.sort_index(inplace=True)
b.head(3)
```

```python
b.columns=[column_name.replace(" ", "_")for column_name in b.columns]
b.head(1)
```

```python
b.query('Actor=="Sean Connery"')
```

```python
b.query('Director=="Terence Young"')
```

```python
b.query("Actor!= 'Roger Moore'")
```

```python
b.query("Actor != 'Roger Moore' and Director=='John Glen'")
```

```python
b.query("Actor in ['Timothy Dalton', 'George Lazenby']")
```

```python
b.query("Actor not in ['Timothy Dalton', 'George Lazenby']")
```

# A review of the .apply() Method on Single Columns

```python
b=pd.read_csv("Datasets/jamesbond.csv", index_col="Film")
b.sort_index(inplace=True)
b.head(3)
```

```python
def conver_to_string_and_add_millions(number):
    return str(number)+" Millions" # Transform a numeric input to a str 

b["Box Office"]=b["Box Office"].apply(conver_to_string_and_add_millions)
```
